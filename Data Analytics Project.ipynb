```python
# Missing values
def missing_values_table(df):
        # Total missing values
        mis_val = df.isnull().sum()
        
        # Percentage of missing values
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        
        # Make a table with the results
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        
        # Rename the columns
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        
        # Sort the table by percentage of missing descending
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        
        # Print some summary information
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
            "There are " + str(mis_val_table_ren_columns.shape[0]) +
              " columns that have missing values.")
        
        # Return the dataframe with missing information
        return mis_val_table_ren_columns
missing_values= missing_values_table(df)
missing_values.style.background_gradient(cmap='Reds')
```

```python
import pandas as pd
from IPython.display import display_html

def max_polluted_city(city_day, pollutant):
    # Ensure input DataFrame is not modified
    df_copy = city_day[[pollutant, 'City']].copy()
    # Calculate mean pollution levels per city
    df_mean = df_copy.groupby('City').mean().sort_values(by=pollutant, ascending=False).reset_index()
    # Round the pollutant values to 2 decimal places
    df_mean[pollutant] = round(df_mean[pollutant], 2)
    # Return the top 10 polluted cities with a background gradient
    return df_mean.head(10).style.background_gradient(cmap='OrRd')

# Display multiple styled DataFrames side by side
def display_side_by_side(*dfs):
    html_str = ''
    for df in dfs:
        html_str += df.to_html()
    display_html(html_str.replace('table', 'table style="display:inline"'), raw=True)

# Assuming city_day is your DataFrame containing pollution data
# Replace this with your actual DataFrame
city_day = pd.DataFrame({
    'City': ['Patna', 'Delhi', 'Gurugram', 'Lucknow', 'Ahmedabad', 'Kolkata', 'Jorapokar', 'BrajrajNagar', 'Guwahati', 'Talcher'],
    'PM2.5': [123.50, 117.20, 117.10, 109.71, 67.85, 64.36, 64.23, 64.06, 63.69, 61.41],
    'PM10': [232.81, 191.50, 165.77, 149.66, 126.75, 124.22, 123.48, 119.32, 116.60, 115.63],
    'NO2': [59.03, 50.79, 40.40, 37.49, 37.19, 33.24, 32.42, 31.35, 28.78, 28.39],
    'SO2': [55.25, 33.65, 28.49, 22.13, 17.60, 15.90, 15.2, 14.66, 14.26, 13.06],
    'CO': [22.19, 2.13, 1.98, 1.85, 1.84, 1.80, 1.63, 1.53, 1.30, 1.26],
    'BTX': [38.23, 37.11, 26.86, 17.43, 15.03, 14.60, 14.58, 10.73, 9.09, 3.68]
})

# Generate the styled DataFrames for each pollutant
pm2_5 = max_polluted_city(city_day, 'PM2.5')
pm10 = max_polluted_city(city_day, 'PM10')
no2 = max_polluted_city(city_day, 'NO2')
so2 = max_polluted_city(city_day, 'SO2')
co = max_polluted_city(city_day, 'CO')
btx = max_polluted_city(city_day, 'BTX')

# Display the styled DataFrames side by side
display_side_by_side(pm2_5, pm10, no2, so2, co, btx)
```
